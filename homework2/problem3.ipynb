```python
import numpy as np
import seaborn as sns
```

```python
### Simulate data
mu = 0
sigma = 1
mu_new = 3
sigma_new = 5
N = 10000
y = np.random.normal(mu, sigma, N)
y_new = sigma_new * y + mu_new
y_new_control = np.random.normal(mu_new, sigma_new, N)
```

```python
sns.histplot(y_new, color="red")
sns.histplot(y_new_control, color="gray")
sns.histplot(y)
```
