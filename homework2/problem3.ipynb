```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mu = 0
sigma = 1
mu_new = 3
sigma_new = 5
num_samples = 10_000

y = np.random.normal(mu, sigma, num_samples)
y_new = sigma_new * y + mu_new
y_new_control = np.random.normal(mu_new, sigma_new, num_samples)
```

```python
sns.histplot(y_new, color="red", alpha=0.5, label="Transformed Data")
sns.histplot(y_new_control, color="gray", alpha=0.5, label="Control Data")
sns.histplot(y, color="blue", alpha=0.5, label="Original Data")

plt.legend()
plt.show()
```
